```python
import pandas as pd
import numpy as np
```

```python
# 데이터 불러오기
df_movie = pd.read_csv(r'./data/kaggle_movie_small/movies.csv')
df_rating = pd.read_csv(r'./data/kaggle_movie_small/ratings.csv')
```

```python
# rating 데이터에 영화이름 가져오기
df_data = pd.merge(df_rating, df_movie, how='inner', on='movieId')
df_data.head()
```

```python
# user Id와 movie Id 행렬을 만들기
df = pd.pivot_table(df_data, index='userId', columns='title', values='rating')
df
```

```python
# Nan값을 0으로 채우기
df = df.fillna(0)
df
```

```python
# 인덱스-컬럼 전치
df_transpose = df.transpose()
```

```python
# COS유사도 구하기 함수
def get_cosine_similarity(u, v):
    mask = np.isfinite(u) & np.isfinite(v)
    
    print( u, v )   # NaN 걸러내기 이전
    
    u = u[mask]
    v = v[mask]
    
    print( u, v )   # NaN 걸러낸 이후
    
    uvdot = (u * v).sum()
    
    norm1 = (u ** 2).sum()
    norm2 = (v ** 2).sum()
    
    score = uvdot / np.sqrt(norm1 * norm2)
    
    return score
```

```python
# 순열 조합하여 COS유사도 구하기
def get_cosine_similarity_table(df_transpose):
    index_combinations = list(product(df_transpose.index, repeat=2))
    
    similarity_list = []
    
    for uname, vname in index_combinations:
        u = df_transpose.loc[uname]
        v = df_transpose.loc[vname]
        
        score = get_cosine_similarity(u, v)
        
        similarity = {
            'u': uname,
            'v': vname,
            'score': score
        }
        
        similarity_list.append(similarity)
        
    similarity_list = pd.DataFrame(similarity_list)
    
    similarity_table = pd.pivot_table(similarity_list, index='u', columns='v', values='score')
    
    return similarity_table
```

```python
# 영화별 COS 유사도
from sklearn.metrics.pairwise import cosine_similarity
item_filtering_COS = cosine_similarity(df_transpose, df_transpose)
item_filtering_COS_df = pd.DataFrame(data=item_filtering_COS, index=df.columns, columns=df.columns)
item_filtering_COS_df
```

```python
# 추천 영화 반환 함수
def find_movie_COS_sim_item(df, title_name, top_n=10):
    COS_sim_movie = df[[title_name]].drop(title_name, axis=0)

    return COS_sim_movie.sort_values(title_name, ascending=False)[:top_n]
```

```python
# 인셉션과 유사한 영화 찾기
find_movie_COS_sim_item(item_filtering_COS_df, 'Inception (2010)')
```

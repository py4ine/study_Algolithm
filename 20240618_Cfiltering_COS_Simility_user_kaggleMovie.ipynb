```python
# COS유사도 구하기 함수
def get_cosine_similarity(u, v):
    mask = np.isfinite(u) & np.isfinite(v)
    
    print( u, v )   # NaN 걸러내기 이전
    
    u = u[mask]
    v = v[mask]
    
    print( u, v )   # NaN 걸러낸 이후
    
    uvdot = (u * v).sum()
    
    norm1 = (u ** 2).sum()
    norm2 = (v ** 2).sum()
    
    score = uvdot / np.sqrt(norm1 * norm2)
    
    return score
```

```python
# 순열 조합하여 COS유사도 구하기
def get_cosine_similarity_table(df_transpose):
    index_combinations = list(product(df_transpose.index, repeat=2))
    
    similarity_list = []
    
    for uname, vname in index_combinations:
        u = df_transpose.loc[uname]
        v = df_transpose.loc[vname]
        
        score = get_cosine_similarity(u, v)
        
        similarity = {
            'u': uname,
            'v': vname,
            'score': score
        }
        
        similarity_list.append(similarity)
        
    similarity_list = pd.DataFrame(similarity_list)
    
    similarity_table = pd.pivot_table(similarity_list, index='u', columns='v', values='score')
    
    return similarity_table
```

```python
def predict_rating(user_name, movie_name):
    
    # 자신제외
    neighbors_df = df[movie_name].drop(index = user_name)
    neighbors_similarity = user_filtering_COS_df[user_name].drop(index = user_name)

    # 
    nominator = (neighbors_similarity * neighbors_df).sum()
    denominator = neighbors_similarity.sum()
    score = nominator / denominator

    return score
```

```python
# 평점 테이블
from itertools import product
print(len(list(product(df.index, df.columns))))
# list(product(df.index, df.columns))
```

```python
def predict_rating_table(ratings):
    user_movie_combinations = list(product(df.index, df.columns))

    rating_list = []

    for user_name, movie_name in user_movie_combinations:
        score = predict_rating(user_name, movie_name)

        rating_predict = {
            'user' : user_name,
            'movie' : movie_name,
            'score' : score
        }

        rating_list.append(rating_predict)
    
    rating_list = pd.DataFrame(rating_list)

    rating_table = pd.pivot_table(rating_list, index='user', columns='movie', values='score')
    rating_table = rating_table[ratings.isnull()]

    return rating_table
```

```python
rating_table = predict_rating_table(df)
rating_table
```

```python
# 사용자 기반 영화 추천 함수
def predict_movie(user_name, k):
    prediction_list = rating_table.loc[user_name].sort_values(ascending=False)
    prediction_list = prediction_list.head(k).index

    return prediction_list
```

```python
# 해당 영화를 추천해줄 사용자 찾기 함수
def predict_user(movie_name, k):
    prediction_list = rating_table[movie_name].sort_values(ascending = False)
    prediction_list = prediction_list.head(k).index

    return prediction_list
```

```python
# 1번 유저에게 추천할 영화
predict_movie(1, k = 2)
```

```python
# 인셉션 영화를 추천해 줄 사용자
predict_user("Inception (2010)", 5)
```
